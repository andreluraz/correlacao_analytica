```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
#importando dataset built in do sklearn "boston house prices"
from sklearn.datasets import load_boston 
```

```python
data_boston = load_boston()
```

```python
dataset_boston = pd.DataFrame(data_boston.data, columns = data_boston.feature_names)
```

```python
dataset_boston['target'] = data_boston.target
```

```python
dataset_boston.head()
```

```python
X_boston = dataset_boston.drop('target', axis=1)
y_boston = dataset_boston['target']
```

```python
X_boston.corr()
```

```python
plt.figure(figsize = (10,6))
sns.heatmap(X_boston.corr())
```

```python
plt.figure(figsize=(10,5))
plt.scatter(X_boston['NOX'], X_boston['INDUS'])
```
